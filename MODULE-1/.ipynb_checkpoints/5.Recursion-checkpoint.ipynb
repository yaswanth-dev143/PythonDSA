```python
def fun():
    print("GFG")
    fun()
fun()
```

```python
def fun(n):
    if n == 0:
        return
    print("Gfg")
    fun(n-1)
fun(4)
```

# 3. PRACTICE IN RECURSION -1

```python
def fun(n):
    if n == 0:
        return
    print(n)
    fun(n-1)
    print(n)
fun(3)
```

```python
def fun(n):
    if n == 0:
        return
    fun(n-1)
    print(n)
    fun(n-1)
fun(3)
```

# 4. PRACTICE IN RECURSION - 2

```python
def fun(n):
    if n <= 1:
        return 0
    else:
        return 1 + fun(n/2)

fun(16)   
```

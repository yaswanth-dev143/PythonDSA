# 3. PRACTICE IN RECURSION -1

```python
def fun(n):
    if n == 0:
        return
    print(n)
    fun(n-1)
    print(n)
fun(3)
```

```python
def fun(n):
    if n == 0:
        return
    fun(n-1)
    print(n)
    fun(n-1)
fun(3)
```

# 4. PRACTICE IN RECURSION - 2

```python
def fun(n):
    if n <= 1:
        return 0
    else:
        return 1 + fun(n/2)

fun(16)   
```

# 8. SUM OF DIGITS

```python
def sumOfnums(n):
    if n < 10:
        return n
    return sumOfnums(n//10)+n%10
sumOfnums(int(input()))
```
